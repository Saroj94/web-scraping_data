```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_manufacturing_companies_by_revenue"
req = requests.get(url)
```

```python
print(req)
```

```python
#beautify the xml
soup=BeautifulSoup(req.text,"html")
```

```python
print(soup)
```

```python
table=soup.find('table', class_ = 'wikitable sortable plainrowheads').tbody
```

```python
table
```

```python
head=table.find_all('th')
```

```python
hnames=[h.text.strip() for h in head]
```

```python
hnames
```

```python
df=pd.DataFrame(columns= hnames)
```

```python
df
```

```python
table
```

```python
rows=table.find_all('tr')
```

```python
for i in rows[1:]:
    tab=i.find_all('td')
    rowdt=[rtx.text.strip().replace('\n','').replace('/\xa0',' ') for rtx in tab]
    l=len(df)
    df.loc[l]=rowdt
```
