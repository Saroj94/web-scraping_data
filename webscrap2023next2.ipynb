```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://pokemon.fandom.com/wiki/List_of_Pokémon_by_evolution'
```

```python
req=requests.get(url)
```

```python
soup=BeautifulSoup(req.text,'lxml')
```

```python
tab=soup.find('div', class_ = "mw-parser-output")
```

```python
head=tab.find_all('th')
```

```python
title=[name.text.strip() for name in head]
```

```python
head=title
```

```python
tab_dat=[]
for row in tab.find_all('tr'):
    row_data = [cell.text.strip() for cell in row.find_all('td') if cell]
    if row_data:
        tab_dat.append(row_data)
```

```python
df=pd.DataFrame(tab_dat)
```

```python
table = tab_dat

# Get the length of the first row
expected_length = len(table[0])

# Create a list to store indices of mismatched rows
mismatched_indices = []

# Iterate through the table and find rows with different lengths
for i, row in enumerate(table):
    if len(row) != expected_length:
        mismatched_indices.append(i)

if mismatched_indices:
    print("Mismatched columns found in the following rows:")
    for index in mismatched_indices:
        print(f"Row {index + 1}: {table[index]}")
else:
    print("No mismatched columns found in the table.")
```
