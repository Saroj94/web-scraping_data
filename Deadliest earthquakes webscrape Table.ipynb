```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://en.wikipedia.org/wiki/Lists_of_21st-century_earthquakes"
```

```python
req=requests.get(url)
print(req)
```

```python
soup=BeautifulSoup(req.text,"html")
```

```python
tab=soup.find('div', class_ = "mw-content-ltr mw-parser-output")
```

```python
head=tab.find_all('table')[3]
```

```python
col_head=head.find_all('th')
```

```python
name=[]
for i in col_head:
    title=i.text.strip()
    name.append(title)
```

```python
name
```

```python
df=pd.DataFrame(columns=name)
```

```python
df
```

```python
row_datas=[]
for tdata in head.find_all('tr'):
    dat=[rows.text.strip() for rows in tdata.find_all('td') if any(rows)]
    row_datas.append(dat)
```

```python
row_datas
```

```python
dfs=pd.DataFrame(row_datas)
```

```python
dfs.columns=name
```

```python
dfs
```

```python
dfs.to_csv('Deadliest earthquake.csv')
```
